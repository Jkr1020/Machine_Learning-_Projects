```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\j.KISHORERAM\Desktop\New folder (2)\delaney_solubility_with_descriptors.csv") 
```

```python
df.head()
```

```python
y=df['logS']
y
```

```python
#Data seperation
x=df.drop('logS', axis=1)
x
```

###  Data splitting

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

 ## Model building

### Linear Regression
##### Training the model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

#### Applying the model to make a prediction

```python
y_lr_train_pred=lr.predict(x_train)
```

```python
y_lr_test_pred=lr.predict(x_test)
```

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

### Evaluate the model


```python
y_lr_train_pred
```

```python
y_train
```

```python
from sklearn.metrics import mean_squared_error,r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)
```

```python
lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
print("Training mean Squared error(LR):",lr_train_mse)
print("Training r2 score(LR):",lr_train_r2)
```

```python
print("Testing mean Squared error:",lr_test_mse)
print("Testing r2 score:",lr_test_r2)
```

```python
lr_results=pd.DataFrame(["Linear regression:",lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=['Method','Train MSE','Train R2','Test MSE','Test R2']
```

```python
lr_results
```

### Random Forest
#### Train the model

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

#### Apply the model to make a prediction

```python
y_rf_train_pred=lr.predict(x_train)
y_rf_test_pred=lr.predict(x_test)
```

#### Evaluation  model performance

```python
rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2=r2_score(y_train,y_rf_train_pred)
```

```python
rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
```

```python
rf_results=pd.DataFrame(['Random Forest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns=['Method','Train MSE','Train R2','Test MSE','Test R2']
```

```python
rf_results
```

#### Model comparison

```python
df_models=pd.concat([lr_results,rf_results],axis=0)
```
